```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Datasets/Training.csv')
df.drop(columns=['Name','Ticket','Fare','Cabin','Embarked','PassengerId'],inplace=True)
```

```python
df
```

```python
df['Sex'] = df['Sex'].replace({
   'male' : 0,
    'female' : 1
})
```

```python
X = df.copy()
X.drop(columns=['Survived'],inplace=True)
Y = df['Survived']
X
```

```python
X['Age'].fillna( X['Age'].mean() ,inplace=True)
```

```python
columns= X.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        X[columns[i]+'_'+columns[j]] = X[columns[i]] * X[columns[j]]
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X)
```

```python
alg = LogisticRegression(max_iter=2000)
alg.fit( scaler.transform(X),Y)
```

```python
X_test = pd.read_csv('Datasets/Testing.csv')
X_test['Sex'] = X_test['Sex'].replace({
   'male' : 0,
    'female' : 1
})
X_test.drop(columns=['Name','Ticket','Fare','Cabin','Embarked','PassengerId'],inplace=True)
X_test['Age'].fillna( X_test['Age'].mean() ,inplace=True)
X_test.describe()
```

```python
columns= X_test.columns
for i in range(len(columns)):
    for j in range(i,len(columns)):
        X_test[columns[i]+'_'+columns[j]] = X_test[columns[i]] * X_test[columns[j]]
```

```python
dataframe = {
    'PassengerId' : pd.read_csv('Datasets/Testing.csv')['PassengerId'] ,
    'Survived' : alg.predict( scaler.transform(X_test))
}
df = pd.DataFrame(dataframe)
df.to_csv('Datasets/Prediction.csv',index=False)
```
